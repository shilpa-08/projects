###  Finding the best RandomState

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
maxaccuracy=0
maxrandomstate=0
for i in range(1,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=i)
    lg=LogisticRegression()
    lg.fit(x_train,y_train)
    predlg=lg.predict(x_test)
    accuracy=accuracy_score(y_test,predlg)
    if accuracy > maxaccuracy :
        maxaccuracy =accuracy
        maxrandomstate = i
print(f'max accuracy score is {maxaccuracy} , and the best random state is {maxrandomstate}')
```

we have seen the best random state is 417 giving accuracy score of 87.8 %

```python
from sklearn.metrics import confusion_matrix, classification_report
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=417)
lg=LogisticRegression()
lg.fit(x_train,y_train)
predlg=lg.predict(x_test)
print(accuracy_score(y_test,predlg))
print(confusion_matrix(y_test,predlg))
print(classification_report(y_test,predlg))
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtc.fit(x_train,y_train)
pred=dtc.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
rfc.fit(x_train,y_train)
pred=rfc.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from sklearn.svm import SVC

svc=SVC()
svc.fit(x_train,y_train)
svc.fit(x_train,y_train)
pred=svc.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

with random forest classifier and LogisticRegression we are getting highest and similar accuracy , so lets go for cross-validation

```python
from sklearn.model_selection  import cross_val_score
scr=cross_val_score(lg,x,y,cv=5)
print("Cross Validation Score of Logistic Regression model is :",scr.mean())
print("\n\n")
scr=cross_val_score(dtc,x,y,cv=5)
print("Cross Validation Score of Decision Tree model is :",scr.mean())
print("\n\n")
scr=cross_val_score(rfc,x,y,cv=5)
print("Cross Validation Score of Random Forest model is :",scr.mean())
print("\n\n")
scr=cross_val_score(svc,x,y,cv=5)
print("Cross Validation Score of Support Vector model is :",scr.mean())
```

```python
# MODEL                    ACCURACY              CROSS_VAL_SCORE                   DIFFERNCE

# Logistic                       87                        79                          8
# DecisionTree                   76                        78                          2
# RandomForest                   87                        83                          4
# SVC                            86                        80                          6
```

the highest accuracy and less differnce if of RandomForest and DecisionTree, so will go for RandomForest
